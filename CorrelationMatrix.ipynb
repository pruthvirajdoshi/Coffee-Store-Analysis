```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from matplotlib.colors import LinearSegmentedColormap

# Import the data
df = pd.read_csv("cleaned_data.csv")

# Calculate the correlation matrix
correlation_matrix = df.corr()

# Define the colors for the custom colormap (yellow for 1, white for 0, green for -1)
colors = ["green", "white", "yellow"]  # Make sure to place the colors in the order for -1, 0, 1

# Create a custom colormap
custom_colormap = LinearSegmentedColormap.from_list("custom_colormap", colors)

# Create a heatmap of the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap=custom_colormap, center=0, fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```
